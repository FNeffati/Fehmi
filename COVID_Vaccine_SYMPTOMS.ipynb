```
import pandas
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

##imported my library and then imported my file ( a file I have curated by collecting large amount of data about patients and their symptoms to a certain vaccine(in this case let's it's covid vaccine))
##Since not enough research is done on the COVID-19 vaccine, we want to allow the people to enter their own data and help make this data accessible to researchers.

file_name = "https://raw.githubusercontent.com/FNeffati/Fehmi/master/dataset.csv"
data = pandas.read_csv(file_name)
 
#Next, for the sake of this assignment, I am going to reduce the amount of columns from 24 thousand to 436 by removing all duplicate symptoms.

data = data.dropna()
data.sort_values("SYMPTOM", inplace = True)
data.drop_duplicates(subset ="SYMPTOM", 
                     keep = False, inplace = True) 
data

#This is going to show us the Age distribution per state as form of a boxplot.

fig = px.box(data, x="STATE", y="AGE_YRS")
fig.show()

```

```
#Now we are going to see what are the symptoms per state look like

import plotly.express as px
df = data
fig = px.pie(df, values='AGE_YRS', names='STATE',
             title='Symptoms By State',
             hover_data=['SYMPTOM'], labels={'SYMPTOM':'Symptom'})
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```
## Now we are going to try to get what is the most common Symptoms overall:
file_name = "https://raw.githubusercontent.com/FNeffati/Fehmi/master/dataset.csv"
c = pandas.read_csv(file_name)
b = c.pivot_table(index=['SYMPTOM'], aggfunc='size')
b.sort_values (inplace = True)

top = b.nlargest(10,'first')


#sns.displot(top, x = b.nlargest(10,'first'), y = 'SYMPTOM', bins = 50)
#plt.figure(figsize=(20,20))



import plotly.express as px
fig = px.scatter(top, x= b.nlargest(10,'first'), y= top.index)

fig.update_layout(title="most common Symptoms",
                  xaxis_title="Frequency",
                  yaxis_title="Symptom")
fig.show()




```
